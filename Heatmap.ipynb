```python
import urllib.request
import pandas as pd
import seaborn as sns
```

```python
source_url = "https://raw.githubusercontent.com/foerstner-lab/Bits_and_pieces_for_the_carpentries_workshops/main/text_file_examples/Expression_comparison.csv"
```

```python
expression_df = pd.read_csv(source_url, sep="\t", comment="#")
```

```python
expression_df
```

```python
expression_df.set_index('Gene', inplace=True)
```

```python
expression_df
```

```python
ax = sns.heatmap(expression_df)
```

```python
ax = sns.heatmap(expression_df, annot=True)
```

```python
ax = sns.heatmap(expression_df, annot=True, cmap="YlGnBu")
```

```python
ax = sns.clustermap(expression_df, annot=True, cmap="YlGnBu")
```

```python
ax.savefig("heatmap.pdf")
```
